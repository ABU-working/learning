```python
#生成验证码 Generate verification code

import random

def generate_code(code_len=4):
    all_chars='0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    select_list=len(all_chars)-1
    code=''
    for _ in range(code_len):
        temp=random.randint(0,select_list)
        code+=all_chars[temp]
    return code


if __name__=='__main__':
    i=input("Please input 'y' for getting a generatecode:")
    if i=="y":
        code=generate_code(code_len=4)
    print(code)
```

```python
#设计一个函数返回给定文件名的后缀名
def get_suffix(filename,has_dot=False):
    
    pos=filename.rfind('.')#rfind() 返回字符串最后一次出现的索引，如果没有匹配项则返回-1
                            #find() 返回字符串第一次出现的索引，如果没有匹配项则返回-1
    if 0<pos<len(filename)-1:
        temp=pos if has_dot else pos+1
        return filename[temp:]
    else:
        return "Please add suffix."
    
if __name__=='__main__':
    i=input('please input filename:')
    print(get_suffix(i,True))
```

```python
from random import randrange, randint, sample


def display(balls):
    """
    输出列表中的双色球号码
    """
    for index, ball in enumerate(balls):#enumerate() 函数用于将一个可遍历的数据对象
        #(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标，一般用在 for 循环当中。
        if index == len(balls) - 1:
            print('|', end=' ')   #末尾加  |
        print('%02d' % ball, end=' ')  #输出两位数的  数字
    print()


def random_select():
    """
    随机选择一组号码
    """
    red_balls = [x for x in range(1, 34)]
    selected_balls = []
    selected_balls = sample(red_balls, 6)#sample(seq, n) 从序列seq中选择n个随机且独立的元素；
    selected_balls.sort()#sort() 函数用于对原列表进行排序，如果指定参数，则使用比较函数指定的比较函数。
                          #list.sort( key=None, reverse=False)
                          # key -- 主要是用来进行比较的元素，只有一个参数，具体的函数的参数就是取自于可迭代对象中，指定可迭代对象中的一个元素来进行排序。
                          # reverse -- 排序规则，reverse = True 降序， reverse = False 升序（默认）。
    selected_balls.append(randint(1, 16))
    return selected_balls


def main():
    n = int(input('机选几注: '))
    for _ in range(n):
        display(random_select())


if __name__ == '__main__':
    main()
```

```python
def main():
    persons = [True] * 30  #30人的数组，有两个值
    counter, index, number = 0, 0, 0# 计数，默认，成员
    while counter < 15:   #小于15，结束信号
        if persons[index]:  #默认成员，这个成员的位置没有被扔出海
            number += 1      #标记后移
            if number == 9:   #计数器，标准达到，仍出海
                persons[index] = False  #标记这个数
                counter += 1      #从0开始，循环加一
                number = 0        #归零计数 
        index += 1           #和上一个计数器一起，在寻找第九个人，不断计数
        index %= 30            #如果  index 是30 他会被整除，变成余数，当小于30，他会变成余数 就是自己！！！这是关键的轮回
    for person in persons:
        print('基' if person else '非', end='')# false的被跳过，true的继续，靠取余来循环，66666  这是赋值


if __name__ == '__main__':
    main()
```

```python
"""

"""
import os


def print_board(board):
    print(board['TL'] + '|' + board['TM'] + '|' + board['TR'])
    print('-+-+-')
    print(board['ML'] + '|' + board['MM'] + '|' + board['MR'])
    print('-+-+-')
    print(board['BL'] + '|' + board['BM'] + '|' + board['BR'])###制作棋盘，分别是9个位置的值，以及边线


def main():
    init_board = {
        'TL': ' ', 'TM': ' ', 'TR': ' ',
        'ML': ' ', 'MM': ' ', 'MR': ' ',
        'BL': ' ', 'BM': ' ', 'BR': ' '
    }#给棋盘赋值，定义每一个位置！！！这是一个位置对应一个赋值，是字典，此时 赋值为空
    begin = True
    while begin:
        curr_board = init_board.copy()#   “=”的赋值是引用，后值改变的同时前值改变 。
# copy是浅复制，复制一级目录，引用二级目录.这里 一级目录 如TL:是不会改变的，而：对应的键值会改变
        begin = False    #游戏已经开始
        turn = 'x'        #x  棋手 就位
        counter = 0        # 计数
        os.system('clear')  #清空屏幕，保证新棋盘正常进行
        print_board(curr_board)##输出棋盘 此时！！！把curr_board 列表的值赋予棋盘各个位置
        while counter < 9:##最多走9步
            move = input('轮到%s走棋, 请输入位置: ' % turn)  #提示棋手走棋
            if curr_board[move] == ' ':    #棋手输入了对应的九宫格，如果是空的
                counter += 1                  #计数+1，既正常下一步棋----->如果这两个人下了同一个格子，此步跳过，并不会交换棋手和继续进行
                curr_board[move] = turn      #在对应的位置改变键值  如  TL:" "  -->  TL:x
                if turn == 'x':           #交换棋手
                    turn = 'o'
                else:
                    turn = 'x'
            os.system('clear')           #将棋盘清空
            print_board(curr_board)     #输出改变值的棋盘 
        choice = input('再玩一局?(yes|no)')  #
        begin = choice == 'yes'  # begin是bool值，这句话的意思是，如果choice=yes，为真，可以重新开始


if __name__ == '__main__':
    main()
```
